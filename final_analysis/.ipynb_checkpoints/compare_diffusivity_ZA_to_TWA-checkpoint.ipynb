## Compare diffusivity tensors

Here we compare some properties of the diffusivity tensors from the ZA and TWA. 

```python
# load modules 

import numpy as np
import xarray as xr
from xmitgcm import open_mdsdataset
from matplotlib import pyplot as plt
from matplotlib import colors, ticker, cm
from matplotlib.colors import LogNorm
%matplotlib inline
from xgcm import Grid
```

```python
diff_tensor_ZA = xr.open_dataset('analysis_data_files/diff_tensor_ZA_50km.nc')
diff_tensor_TWA = xr.open_dataset('analysis_data_files/diff_tensor_TWA_50km_av800.nc')
```

```python
zeta_bar = xr.open_dataset('analysis_data_files/zeta_bar_800.nc')
zeta_bar_coarse = zeta_bar.zeta.coarsen(XC=10, YC=10).mean()
```

```python
Tcenter = np.linspace(0,8, 101)
dT = Tcenter[1] - Tcenter[0]
Touter = np.append(Tcenter - dT/2, Tcenter[-1] + dT/2)
```

```python
diff_tensor_TWA = diff_tensor_TWA.assign_coords({'Tc': Touter})

grid_iso = Grid(diff_tensor_TWA, coords = {'T':{'center':'T','outer':'Tc'}},
               periodic='X')
```

```python
Zcenter = diff_tensor_ZA.Z.sel(Z=slice(-200, -3000))
```

```python
temp =  grid_iso.transform(diff_tensor_TWA.eigvalsSTcorr, 'T', Zcenter,
                          target_data=zeta_bar_coarse, method='linear')
```

```python
temp
```

```python
temp.sel(ii=0).sel(Z=-1000, method='nearest').plot(vmin=-15e3)
```

```python
diff_tensor_ZA.eigvalsSTcorr.sel(ii=1).sel(Z=-1000, method='nearest').plot(vmin=-15e3)
```

```python
plt.scatter(diff_tensor_ZA.eigvalsSTcorr.sel(ii=1).sel(Z=-1000, method='nearest'),
           temp.sel(ii=0).sel(Z=-1000, method='nearest'), 
           s=0.4)
plt.xlim([-5e3, 15e3])
plt.ylim([-5e3, 15e3])
plt.grid()
```

```python
plt.scatter(diff_tensor_ZA.eigvalsSTcorr.sel(ii=1).sel(Z=-400, method='nearest'),
           temp.sel(ii=0).sel(Z=-400, method='nearest'), 
           s=0.4)
plt.xlim([-5e3, 15e3])
plt.ylim([-5e3, 15e3])
plt.grid()
```

```python
plt.scatter(diff_tensor_ZA.eigvalsSTcorr.sel(ii=1).sel(Z=-1800, method='nearest'),
           temp.sel(ii=0).sel(Z=-1800, method='nearest'), 
           s=0.4)
plt.xlim([-5e3, 15e3])
plt.ylim([-5e3, 15e3])
plt.grid()
```
